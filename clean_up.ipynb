## Найдены пропущенные значения

В датасете пропуски есть только в трех колонках:

- `price_metric` - около 1.16%;
- `address` - около 0.55%;
- `metro` - около 0.02%.

Критически важных пропусков в основных полях нет: `price`, `title`, `id`, `source`, `person_type`, `publication_time` заполнены полностью.

Колонка `price_metric` не будет использоваться как исходный признак, потому что цену за квадратный метр надежнее рассчитать самостоятельно после извлечения площади из `title`.

Пропуски в `address` и `metro` составляют малую долю данных. Чтобы не терять наблюдения, значения будут заполнены категорией `Unknown`.

```python
df_clean = df.copy()

df_clean["address"] = df_clean["address"].fillna("Unknown")
df_clean["metro"] = df_clean["metro"].fillna("Unknown")

df_clean = df_clean.drop(columns=["price_metric"])
```
